```python
import os
import re
```

```python
#Download the list of commanders again
os.system('cd scraper/mtgtop8; scrapy crawl mtgtop8Search -O scraps/mtgtop8_commanders.json')
```

```python
commander_name = "Juri, Master Of The Revue"
commander_file = re.sub("[ ,/]", "_", commander_name)
```

```python
#Scrap decklist for given commander link in scraps/<commander_name>_links.json file (--)
os.system('cd scraper/mtgtop8; scrapy crawl mtgtop8SearchParams -O scraps/'+ commander_file +'_links.json -a commander_name="'+ commander_name +'"')
```

```python
#Scrap decklist for given commander link in scraps/<commander_name>_decklist.json file (--)
os.system('cd scraper/mtgtop8; scrapy crawl mtgtop8Deck -O scraps/'+ commander_file +'_decklist.json -a commander_name="'+ commander_name +'"')
```
