```python
from requests import get
from json import loads

FILE_PATH = "Listes/"
```

```python
deck_id = "uJ7VvgMPRkWdqqskhJIEMA"
```

```python
deck_base_url = "https://api2.moxfield.com/v2/decks/all/%s"
download_base_url = "https://api.moxfield.com/v1/decks/all/%s/download?exportId=%s&arenaOnly=false"
deck_url =  deck_base_url % deck_id
```
